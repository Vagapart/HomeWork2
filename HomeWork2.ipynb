### *Задание 3

```python
x = np.linspace(-3, 3, 51)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 8)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График х**2')
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title('График 2 * х + 0.5')
ax3.plot(x, y3)
ax3.set_title('График -3 * x - 1.5')
ax4.plot(x, y4)
ax4.set_title('Синус')
```

```python
import pandas as pd
##plt.style.use('fivethirtyeight')
from matplotlib import pyplot
```

```python
pdset = pd.read_csv('creditcard.csv\creditcard.csv')  
```

```python
df = pdset['Class'].value_counts()
```

```python
plt.plot(df)
plt.show()
```

```python
df.plot()
plt.yscale(value="log")

plt.show()
```

```python
pdsetClass1 = pdset[pdset['Class'] == 1]
pdsetClass0 = pdset[pdset['Class'] == 0]


pyplot.hist(pdsetClass1['V1'], bins = 20, alpha=0.5, label='Class 1', color = 'red')
pyplot.hist(pdsetClass0['V1'], bins = 20, alpha=0.5, label='Class 0', color = 'grey')
pyplot.legend(loc='upper left')
pyplot.xlabel("V1")

pyplot.show()
```
